```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

class DataCenterEnv:
    def __init__(self):
        self.temp = 25.0
        self.history = []
        
    def step(self, setpoint):
        # Simple thermal model
        self.temp = self.temp * 0.9 + setpoint * 0.1 + np.random.normal(0, 0.1)
        self.history.append(self.temp)
        return self.temp

env = DataCenterEnv()
for _ in range(50):
    env.step(22.0)

plt.plot(env.history)
plt.title('Data Center Thermal Profile (OpenEnv)')
plt.xlabel('Step')
plt.ylabel('Temperature (C)')
plt.show()
```
